```python
## make figure and subplot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 5))
fig.subplots_adjust(wspace=0)
```

```python
## pie chart Data
plt.rcParams['font.family'] = 'NanumGothic' # 폰트 설정
folder = ['맥주','와인','소주','막걸리','위스키','전통주','지역소주','고량주','럼','보드카']

folder1 = ['맥주', '와인','소주','막걸리']
folder2 = ['위스키','전통주','지역소주','고량주','럼','보드카']
# years = ['2019','2020']
explode=[]
df = pd.DataFrame()
df2 = pd.DataFrame()
for a,b,c in zip(folder,folder1,folder2):
    data = pd.read_csv('data/Filter_Data/네이버%s뉴스_2020.csv'%(str(a)))
    data_groupby = data.groupby(['yyyymm']).agg('size')
    s = data_groupby.iloc[0:1]
    new_Data = pd.DataFrame(s)
    if a==b:
        df = pd.concat([df,new_Data])
    else:
        df2 = pd.concat([df2,new_Data])
df['name'] = folder1
    

# df['name'] = folder
# df
# angle = -180 * df[0]

ax1.pie(x=list(df[0]), 
        autopct='%1.1f%%', 
        # startangle=angle, 
        labels=folder1)
        # explode=explode)

plt.show()
```
