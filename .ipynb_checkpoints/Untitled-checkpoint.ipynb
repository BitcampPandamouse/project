```python
import pandas as pd

```

```python
df = pd.read_csv('네이버주류시장뉴스test.csv')


```

```python
df
```

```python
# 1. colums 수정
df.columns=['num','년월일','제목','본문']
df
```

```python
# 2. 중복값 제거
df = df.drop_duplicates(subset=None,keep='first',inplace=False)
df.info()
df

```

```python
# 3. 그룹화(날짜별)
df_groupby = df.groupby(['년월일']).size()
df_groupby

```

```python
df_groupby.plot()


```

```python

# 저장
df.to_csv('네이버뉴스결과피터링.csv', index=False)
```
